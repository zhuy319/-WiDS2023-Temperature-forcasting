```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import FunctionTransformer
```

```python
train = pd.read_csv('/content/drive/Shareddrives/Kaggle/Kaggle/Data/trainknn_all.csv')
test = pd.read_csv('/content/drive/Shareddrives/Kaggle/Kaggle/Data/test_all.csv')
```

```python
target = 'contest-tmp2m-14d__tmp2m'
X_train = train.drop([target], axis=1)
features = X_train.columns
X_test = test[features]
y_train = train[target]
```

## LightGBM

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error as mse
import lightgbm as lgbm
from tqdm import tqdm
```

```python
params= {
  'learning_rate':0.05, # 0.05/0.1 no big diff
  'num_leaves': 16,
  'feature_fraction': 0.5,
  'bagging_fraction': 0.9,
  'bagging_freq': 4,
  'min_data_in_leaf': 1000,
  'max_depth': 8,
  'num_iterations':1000 # 1000/500 no big diff
}

model = lgbm.LGBMRegressor(**params)    

model.fit(X= X_train,
              y= y_train
             )
preds = model.predict(X_test)
```

```python
sample = pd.read_csv('/content/drive/Shareddrives/Kaggle/Kaggle/Output/sample_solution.csv')
preds1 = sample.copy()
preds1['contest-tmp2m-14d__tmp2m'] = preds
with open('/content/drive/Shareddrives/Kaggle/Kaggle/Output/solution_v19.csv', 'w', encoding = 'utf-8-sig') as f:
    preds1.to_csv(f,index=False)
```

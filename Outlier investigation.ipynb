```python
train = pd.read_csv('/content/drive/Shareddrives/Kaggle/Kaggle/Data/trainknn_all.csv')
```

```python
def extract_outliers(dataset, num_std):
   numeric_columns = dataset.select_dtypes('float64').columns
   for column in numeric_columns: 
      mean = dataset[column].mean()
      sd = dataset[column].std() 
      outliers = dataset[(dataset[column] <= mean + (num_std * sd))]
      outliers = dataset[(dataset[column] >= mean - (num_std * sd))]
   return outliers
```

```python
outliers = extract_outliers(train,5)
outliers.shape
```

```python
fig = plt.figure(figsize=(26, 20))
plt.boxplot(outliers)
plt.show()
```
